```python
#converting certain columns to numeric type
Pesticides[['UN_NUMBER', 'LICENCE_DATE', 'PEST_KEY_CODE']] = Pesticides[['UN_NUMBER', 'LICENCE_DATE', 'PEST_KEY_CODE']].apply(pd.to_numeric, errors='coerce')
Pesticides.dtypes
```

```python
#check for ENVIRONMENT TOXICITY en unique values 
Pesticides["ENVIRONMENT TOXICITY en"].unique()
```

```python
# create a list of all of the ENVIRONMENT TOXICITY that effect bees

bee = ['HIGHLY TOXIC TO BEES',
       'SLIGHTLY TOXIC TO BEES',
       'MODERATELY TOXIC TO BEES',
       'BEES', 'FISH,BEES & SEA CREATURES', 'FISH & BEES',
       'FISH, BEES & BIRDS', 'BEES & SEA CREATURES',
       'BIRDS & BEES']


#filtring the data, aggragation and counting
pesticides_subset = Pesticides.loc[
    (Pesticides["CROP GROUP he"] == "פרחים") &
    (Pesticides["ENVIRONMENT TOXICITY en"].isin(bee))
].groupby(["PRACTICE TYPE","CROP GROUP he","ENVIRONMENT TOXICITY en", "PEST LATIN","PEST_KEY_CODE"]).size().reset_index(name="count")

pesticides_subset
```

```python
#converting a column to numeric type
Pests['PEST_KEY_CODE'] = Pests['PEST_KEY_CODE'].apply(pd.to_numeric, errors='coerce')
Pests.dtypes
```

```python
#merge both data sets based on common culomn

merged = pesticides_subset.merge(Pests[["COMMON_NAMES_IN_HEBREW",
                                        "PESTICIDES_CODE","GENUS","SPECIES","ORDER",
                                        "CLASS", "CODE_NEGA", "PEST_KEY_CODE"
                                       ]].rename(columns={"PEST_KEY_CODE": "PEST_KEY_CODE_Pests"}),
                                  left_on="PEST_KEY_CODE", right_on="PEST_KEY_CODE_Pests")
```
